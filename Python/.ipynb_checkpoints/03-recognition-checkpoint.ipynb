```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from urllib.request import urlopen
import requests
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO

from scipy import stats
from sklearn.ensemble import RandomForestClassifier 
```

### Featch data

```python
link = "https://raw.githubusercontent.com/shuailab/ind_498/master/resource/data/AD.csv"
r = requests.get(link)
data = pd.read_csv(StringIO(r.text))
```

```python
X = data.iloc[:, 1:16]
Y = data[['DX_bl']]

# set training and testing data
x_train, x_test, y_train, y_test = train_test_split(
            X, Y, test_size=0.5)
```

```python
print(x_train.shape)
print(x_test.shape)
```

### Build models

```python
# from sklearn.linear_model import LogisticRegression
# from sklearn import metrics
# logreg = LogisticRegression()
# logreg.fit(x_train, y_train.values.ravel())
# y_pred = logreg.predict(x_test)
 
# print(confusion_matrix(y_pred, y_test))
```

```python
# Build a generalized linear model with all predictor variables
import statsmodels.api as sm
GLM = sm.GLM(y_train, sm.add_constant(x_train), family=sm.families.Binomial())
logit_AD_full = GLM.fit()
print(logit_AD_full.summary())

dev_p_val = 1 - stats.chi2.sf(logit_AD_full.deviance, logit_AD_full.df_resid)
print("p-value: {pvalue}".format(pvalue=dev_p_val))


```

```python

# first use regularization to delete irrelevant or redundant features
# then refit a GLM model
logit_AD_reduced = GLM.fit_regularized(alpha=0.005,refit=True) 
print(logit_AD_reduced.summary())
dev_p_val = 1 - stats.chi2.sf(logit_AD_reduced.deviance, logit_AD_reduced.df_resid)
print("p-value: {pvalue}".format(pvalue=dev_p_val))

```

### Model validation

```python

from sklearn.metrics import confusion_matrix
y_score = logit_AD_reduced.predict(sm.add_constant(x_test))
y_pred = [ 0 if x < 0.5 else 1 for x in y_score]
print("Confusion matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

# Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_score)
plt.plot(fpr, tpr, color = 'red', label='ROC curve')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--') # random prediction
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
```

```python
# get confidence interval of a prediction
prediction = logit_AD_reduced.get_prediction(sm.add_constant(x_test))
pred_ci = prediction.summary_frame()
pred_ci.head(5)
```

```python
# Odds Ratio
np.exp(pred_ci[['mean','mean_ci_lower','mean_ci_upper']]).head(5)
```

#### Monitoring

```python
# Generate data
reference_data = pd.DataFrame( np.random.normal(0, 1, size=[100, 2]),
                    columns=['X1','X2'])
normal_data = pd.DataFrame( np.random.normal(0, 1, size=[100, 2]),
                    columns=['X1','X2'])
abnormal_data = pd.DataFrame({
    'X1': np.random.normal(0, 1, 100),
    'X2': np.random.normal(2, 1, 100)
    })
```

```python
normal = plt.scatter(normal_data.X1, normal_data.X2, marker='o')
abnormal = plt.scatter(abnormal_data.X1, abnormal_data.X2, marker='o')
plt.ylim(-3, 6)
plt.legend((normal, abnormal),
           ('normal', 'abnormal'),
           scatterpoints=1,
           loc='upper right',
           ncol=1,
           fontsize=12)
```

```python
reference_data['label'] = 0
normal_data['label'] = 1
abnormal_data['label'] = 1
test_data = normal_data.append(abnormal_data)
print(test_data.shape)
wsz = 20
score_list = list()
importance_df = pd.DataFrame()

for i in range(len(test_data)-wsz+1):
    this_ix = list(range(i,i+wsz))
    combined_data = reference_data.append( test_data.iloc[this_ix] )
    
    forest = RandomForestClassifier(n_estimators = 100, oob_score = True, class_weight="balanced")
    forest.fit(combined_data[['X1','X2']], combined_data.label)
    score_list.append(forest.oob_score_)
    
    importances = forest.feature_importances_
    importances
    
    this_df = pd.DataFrame({"X1": [importances[0]],
                            "X2": [importances[1]]
                           })
    importance_df = importance_df.append(this_df)
```

```python

from matplotlib import pyplot as plt
# monitoring statistics
plt.plot(score_list)
plt.title('monitoring statistics')
plt.show()

#importance score
x1_imp = plt.plot(importance_df.X1.values,label="X1")
x2_imp = plt.plot(importance_df.X2.values,label="X2")
plt.legend(loc="upper left")
plt.title('monitoring diagnosis')
plt.show()
```

#### Decision tree decision boundry 

```python
# Generate data
data = pd.DataFrame( np.random.uniform(0, 1, size=[4000, 2]),
                    columns=['X1','X2'])
data['X12'] = 0.5 * (data.X1 - data.X2)
data['label'] = 1
data.loc[data.X12>0,'label'] = 0

plt.scatter(data.X1, data.X2, s=1, marker='o', c = data.label)
```

```python
# set training and testing data
x_train, x_test, y_train, y_test = train_test_split(
            data[['X1','X2']], data['label'], test_size=0.8)
```

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression().fit(x_train, y_train)
x_test['pred_lm'] = lm.predict(x_test[['X1','X2']])
plt.scatter(x_test.X1, x_test.X2, s=1, marker='o', c = x_test.pred_lm)
```

```python
from sklearn import tree # import DecisionTreeClassifier
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(x_train, y_train)
x_test['pred_tree'] = clf.predict(x_test[['X1','X2']])
plt.scatter(x_test.X1, x_test.X2, s=1, marker='o', c = x_test.pred_tree)
```

```python
fig, ax = plt.subplots(figsize=(18, 10))  # whatever size you want
tree.plot_tree(clf, max_depth=4, fontsize=12, ax=ax)
plt.show()
```
