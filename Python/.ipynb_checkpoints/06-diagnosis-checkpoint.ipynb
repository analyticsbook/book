```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from urllib.request import urlopen
import requests
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
try:
    from StringIO import StringIO
except ImportError:
    from io import StringIO

from scipy import stats
from sklearn.ensemble import RandomForestClassifier 
```

### Get data

```python
link = "https://raw.githubusercontent.com/shuailab/ind_498/master/resource/data/AD.csv"
r = requests.get(link)
data = pd.read_csv(StringIO(r.text))
data.head(2)
```

### Evaluate models

```python
import statsmodels.api as sm

X = data.iloc[:, :16]
Y = data['MMSCORE']
model = sm.OLS(Y, sm.add_constant( X ) )
model_fit = model.fit()
print(model_fit.summary())
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(model_fit, ax=ax, criterion="cooks")
```

```python
pred = model_fit.fittedvalues.copy()
residual = Y - pred
fig, ax = plt.subplots()
_ = ax.scatter(residual, pred)
```

```python
import scipy as sp
fig, ax = plt.subplots()
_, (__, ___, r) = sp.stats.probplot(residual, plot=ax, fit=True)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model_fit, "DX_bl", fig=fig)
```

### Clustering

```python
# Generate data
cluster1 = pd.DataFrame({
    'X': np.random.normal(0, 1, 200),
    'Y': np.random.normal(0, 1, 200)})
cluster2 = pd.DataFrame({
    'X': np.random.normal(10, 1, 200),
    'Y': np.random.normal(0, 1, 200)})
cluster3 = pd.DataFrame({
    'X': np.random.normal(20, 1, 200),
    'Y': np.random.normal(0, 1, 200)})
cluster4 = pd.DataFrame({
    'X': np.random.normal(40, 2, 200),
    'Y': np.random.normal(0, 1, 200)})

clusters = pd.concat([cluster1, cluster2, cluster3, cluster4], ignore_index=True)
plt.scatter(clusters.X, clusters.Y, s=8, marker='o')
```

```python
# K-means clustering
from sklearn.cluster import KMeans
kmeans = KMeans(init='k-means++', n_clusters=4, n_init=10)
kmeans_fit = kmeans.fit(clusters)
```

```python
clusters['cluster'] = kmeans.predict(clusters)
```

```python
plt.scatter(clusters.X, clusters.Y, s=8, marker='o', c = clusters.cluster)
```

```python
# Gaussian Mixture Model
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4).fit(clusters)
labels = gmm.predict(clusters)

plt.scatter(clusters.X, clusters.Y, s=8, marker='o', c = labels)
```
